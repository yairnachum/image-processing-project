# Week 8 — Distorted stage curves

Loads the three sweep CSVs produced by `scripts/eval_distorted` and renders the three headline SNR-vs-metric curves:

- mean **mAP@0.5** across 15 DOTA classes vs. mean SNR per (distortion, level)
- mean **HED ODS F-score** across 40 tiles vs. mean SNR
- mean **ORB good-match ratio** across 40 tiles vs. mean SNR

Pre-requisite: run `python -m scripts.eval_distorted ...` from the repo root before opening this notebook.

```python
from pathlib import Path
import pandas as pd
from src.figures import plot_metric_vs_snr

REPO = Path.cwd().parent
SWEEP = REPO / "results" / "distortion_sweep"
OUTFIG = REPO / "outputs" / "figures"
OUTFIG.mkdir(parents=True, exist_ok=True)

pc = pd.read_csv(SWEEP / "perclass_detections.csv", dtype={"level": str})
em = pd.read_csv(SWEEP / "edge_metrics.csv",       dtype={"level": str})
om = pd.read_csv(SWEEP / "orb_match.csv",          dtype={"level": str})
print(f"perclass_detections: {len(pc)} rows  |  edge_metrics: {len(em)} rows  |  orb_match: {len(om)} rows")
```

```python
# mAP@0.5 vs SNR — mean over 15 DOTA classes per (distortion, level)
mAP_by_combo = pc.groupby(["distortion", "level", "snr_db_mean"], as_index=False)["ap_iou50"].mean()
plot_metric_vs_snr(
    mAP_by_combo, value_col="ap_iou50",
    title="mAP@0.5 vs SNR (mean over 15 DOTA classes)",
    ylabel="mAP@0.5",
    out_path=OUTFIG / "distorted_curve_map_vs_snr.png",
)
mAP_by_combo.sort_values(["distortion", "snr_db_mean"])
```

```python
# Edge F-score vs SNR — mean over 40 tiles per (distortion, level)
edge_by_combo = em.groupby(["distortion", "level", "snr_db_mean"], as_index=False)["ods_f_score"].mean()
plot_metric_vs_snr(
    edge_by_combo, value_col="ods_f_score",
    title="HED ODS F-score vs SNR (mean over 40 tiles)",
    ylabel="F-score (ODS)",
    out_path=OUTFIG / "distorted_curve_edgeF_vs_snr.png",
)
edge_by_combo.sort_values(["distortion", "snr_db_mean"])
```

```python
# ORB good-match ratio vs SNR — mean over 40 tiles per (distortion, level)
orb_by_combo = om.groupby(["distortion", "level", "snr_db_mean"], as_index=False)["good_ratio"].mean()
plot_metric_vs_snr(
    orb_by_combo, value_col="good_ratio",
    title="ORB good-match ratio vs SNR (mean over 40 tiles)",
    ylabel="good_ratio",
    out_path=OUTFIG / "distorted_curve_orb_vs_snr.png",
)
orb_by_combo.sort_values(["distortion", "snr_db_mean"])
```
